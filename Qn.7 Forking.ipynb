```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = r"C:\Users\Administrator\Documents\GitHub\Forking_Assignment\1500000 Sales Records(1).csv" # Update path accordingly
df = pd.read_csv(file_path)
df.head(3)
```

```python
#Distribution of revenue by order priority-bar chart + pie chart
```

```python
plt.figure(figsize=(8, 6))
df.groupby('Order Priority')['Total Revenue'].sum().sort_values().plot(kind='bar', color='pink')
plt.title('Total Revenue by Order Priority')
plt.xlabel('Order Priority')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
#pie chart 
plt.figure(figsize=(6, 6))
df['Order Priority'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue','green', 'purple', 'orange'])
plt.title('Proportion of Total Revenue by Order Priority')
plt.ylabel('')  # Remove y-axis label
plt.show()
```
