```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('1500000 Sales RecordS(1).csv', parse_dates=['Order Date'])
```

```python
df.head(3)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df.head(3)
```

```python
df = df.sort_values('Order Date')
```

```python
plt.figure(figsize=(10, 6))
df.groupby(df['Order Date'].dt.year)['Total Profit'].sum().plot(kind='line', marker='o', color='green')
plt.title('Total Profit by Date')
plt.xlabel('Order Date')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```

```python
df.groupby(df['Order Date'].dt.to_period('M'))['Total Profit'].sum().plot(kind='line', marker='o', color='blue')
```
