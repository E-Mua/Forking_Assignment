```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = r"C:\Users\k00011888\Documents\GitHub\Forking_Assignment\1500000 Sales Records(1).csv"
fork = pd.read_csv(file_path) 

if "Units Sold" in fork.columns:
    # Create the density plot
    plt.figure(figsize=(8, 5))
    sns.kdeplot(fork["Units Sold"], fill=True, color="blue")

    # Add labels and title
    plt.xlabel("Units Sold")
    plt.ylabel("Density")
    plt.title("Density Plot of Units Sold")

    # Show the plot
    plt.show()
else:
    print("Column 'Units Sold' not found in dataset.")
```

```python
if "Units Sold" in fork.columns:
    # Create the histogram
    plt.figure(figsize=(8, 5))
    sns.histplot(fork["Units Sold"], bins=30, kde=False, color="blue")

    # Add labels and title
    plt.xlabel("Units Sold")
    plt.ylabel("Frequency")
    plt.title("Histogram of Units Sold")

    # Show the plot
    plt.show()
else:
    print("Column 'Units Sold' not found in dataset.")

```
