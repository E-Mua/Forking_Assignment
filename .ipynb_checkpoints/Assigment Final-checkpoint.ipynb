```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path= r"C:\Users\k00009513\Desktop\Forking_Assignment\1500000 Sales Records(1).csv"
data=pd.read_csv(path)
data.head(5)
```

```python
#pie chart 
plt.figure(figsize=(6, 6))
data['Order Priority'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['purple','grey', 'skyblue', 'coral'])
plt.title('Proportion of Total Revenue by Order Priority')
plt.ylabel('')  # Remove y-axis label
plt.show()
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data.head(5)
```

```python
data = data.sort_values('Order Date')
```

```python
plt.figure(figsize=(10, 6))
data.groupby(data['Order Date'].dt.year)['Units Sold'].sum().plot(kind='line', marker='o', color='green')
plt.title('Units Sold by Date')
plt.xlabel('Order Date')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```

```python
data.groupby(data['Order Date'].dt.to_period('M'))['Total Profit'].sum().plot(kind='line', marker='o', color='blue')
```

```python
data.groupby(data['Order Date'].dt.to_period('M'))['Units Sold'].sum().plot(kind='line', marker='o', color='blue')

```
